```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load data into dataframe
achievements = pd.read_csv("../data/schulich_achievements.csv")
achievements.head(20)
```

```python
#Convert the data to a format accepted by matplotlib (for bar and line graphs)
year = achievements['year']
chairs = achievements['chairs_professorship'].tolist()
faculty = achievements['faculty_achievement'].tolist()
student = achievements['student_achievement'].tolist()
alumni = achievements['alumni_achievement'].tolist()
initiative = achievements['initiatives'].tolist()
```

```python
#Stacked bar chart showing total number achievements

plt.figure(figsize=(15,10))
width = .8  
N = len(year)

#Define bars
plt.bar(year, initiative, width, color='violet')
plt.bar(year, alumni, width, color='orange', bottom=initiative)
plt.bar(year, student, width, color='red', bottom=np.array(initiative)+np.array(alumni))
plt.bar(year, faculty, width, color='green', bottom=np.array(initiative)+np.array(alumni)+np.array(student))
plt.bar(year, chairs, width, color='blue', bottom=np.array(initiative)+np.array(alumni)+np.array(student)+np.array(faculty))

#Plot
plt.ylabel('Total Number of Achievments')
plt.title('Schulich Innovations and Recognition')
plt.xticks(year)
plt.legend(['Initiatives','Alumni','Students','Faculty','Chairs/Professorship'])
plt.xticks(rotation=70)

plt.show()
```

```python
#Line graph showing achievements by category

plt.figure(figsize=(15,10))

#Define categories
plt.plot(year, chairs, color='blue')
plt.plot(year, faculty, color='green')
plt.plot(year, student, color='red')
plt.plot(year, alumni, color='orange')
plt.plot(year, initiative, color='violet')

#Plot
plt.xlabel('Year')
plt.ylabel('Number of Achievements')
plt.title('Schulich Innovations and Recognition - Per Category')
plt.legend(['Chairs/Professorship','Faculty','Students','Alumni','Initiatives'])
plt.xticks(rotation=70)
plt.xticks(year)

plt.show()
```

```python
#Alternative to Stacked Bar Chart: Stacked Area Chart

#Convert data to a format accepted by matplotlib
ach_data = achievements[[
     'chairs_professorship', 
     'faculty_achievement', 
     'student_achievement', 
     'alumni_achievement', 
     'initiatives']
].transpose().values.tolist()
 
# Data
x=achievements['year'].tolist()
y=ach_data

# Plot
plt.figure(figsize=(15,10))
plt.stackplot(x,y, labels=['Chairs/Professorships','Faculty Achievement','Student Achievement', 'Alumni Achievement', 'Initiatives'])
plt.legend(loc='upper left')
plt.xticks(rotation=70)
plt.xticks(year)
plt.ylabel('Total Number of Achievments')
plt.title('Schulich Innovations and Recognition')

plt.show()
```
